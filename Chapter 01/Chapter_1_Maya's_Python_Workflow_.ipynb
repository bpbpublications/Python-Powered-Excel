```python
import pandas as pd
import matplotlib.pyplot as plt
import xlsxwriter
```

# Reading and Combining Data Sets





```python
# List of month file names (without extension)
months = ["YT_January", "YT_February", "YT_March", "YT_April", "YT_May", "YT_June"]

# Base GitHub URL

base_url = "https://raw.githubusercontent.com/bpbpublications/Python-Powered-Excel/main/Data/YT_month_wise/"

# Read and concat all files
df_all = pd.concat(
    [
        pd.read_excel(f"{base_url}{m}.xlsx")
        for m in months
    ],
    ignore_index=True
)
```

```python
month_order = ['January', 'February', 'March', "April", "May", "June"]

df_all["month"] = pd.Categorical(df_all["month"],
                                 categories=month_order, ordered=True)

df_all
```

# Computing KPIs

```python
df_all = (df_all
          .assign(
              watch_per_view = lambda df: df["watch_min"] / df["views"],
              click_thr_rate = lambda df: df["views"] / df["impressions"],
              subs_per_views = lambda df: df['subs_net']/df['views'],
              subs_per_imp = lambda df: df['subs_net']/df['impressions']
              ))

df_all
```

```python
monthly_kpis = (df_all
                .groupby('month', observed=True)
                .agg(
                    total_views=("views", "sum"),
                    total_watch_min=("watch_min", "sum"),
                    avg_watch_per_view = ("watch_per_view", lambda x: round(x.mean(), 2)),
                    avg_duration_min=("duration_min", lambda x: round(x.mean(), 2)),
                    total_impressions=("impressions", "sum"),
                    avg_ctr = ("click_thr_rate", lambda x: round(x.mean(), 2)),
                    total_subs=("subs_net", "sum"),
                    total_subs_per_view = ('subs_per_views', 'sum'),
                    total_subs_per_imp = ('subs_per_imp', 'sum')
                )
                .reset_index())

monthly_kpis
```

```python
playlist_kpis = (df_all
                .groupby('playlist', observed=True)
                .agg(
                    total_views=("views", "sum"),
                    total_watch_min=("watch_min", "sum"),
                    total_impressions=("impressions", "sum"),
                    avg_ctr = ("click_thr_rate", lambda x: round(x.mean(), 2))
                )
                .reset_index())

playlist_kpis
```

# Creating Excel Report

```python
last_month = df_all['month'].unique()[-1]
output_file = f"YouTube_Report_upto_{last_month}.xlsx"

with pd.ExcelWriter(output_file, engine="xlsxwriter") as writer:

    wb  = writer.book
    # Formats
    header_format = wb.add_format({"bold": True, "bg_color": "#D7E4BC"})
    title_format = wb.add_format({"bold": True, "font_size": 14, "align": "center", "valign": "vcenter", "bg_color": "#E0EBF5"})

    # --- Combined Data sheet ---
    df_all.to_excel(writer, sheet_name="Combined Data", index=False)

    ws_data = writer.sheets["Combined Data"]

    for col_num, value in enumerate(df_all.columns.values):
        ws_data.write(0, col_num, value, header_format)


    # --- KPIs sheet ---
    ws_kpis = writer.sheets['KPIs'] if 'KPIs' in writer.sheets else wb.add_worksheet('KPIs')

    ws_kpis.merge_range('A1:F1', 'Overall KPIs', title_format)

    # Calculate Overall KPIs
    overall_total_views = df_all['views'].sum()
    overall_avg_ctr = round(df_all['click_thr_rate'].mean(), 2)
    overall_avg_view_duration = round(df_all['duration_min'].mean(), 2)

    # Identify Top Performing Playlist
    top_playlist_row = playlist_kpis.loc[playlist_kpis['total_views'].idxmax()]
    top_playlist_name = top_playlist_row['playlist']
    top_playlist_views = top_playlist_row['total_views']
    top_playlist_watchtime = top_playlist_row['total_watch_min']

    headers = ['Total Views', 'Average CTR %', 'Average View Duration',
               'Top Performing Playlist', 'Top Playlist Views', 'Top Playlist Watchtime']
    ws_kpis.write_row(2, 0, headers, header_format)

    overall_kpis_values = [
        overall_total_views,
        overall_avg_ctr,
        overall_avg_view_duration,
        top_playlist_name,
        top_playlist_views,
        top_playlist_watchtime
    ]
    ws_kpis.write_row(3, 0, overall_kpis_values)

    # --- Monthly KPIs Table ---
    # Write title for Monthly KPIs
    ws_kpis.merge_range('A6:J6', 'Monthly KPIs', title_format)

    # Write monthly_kpis starting from row 8 (0-indexed 7) and col 1 (0-indexed 0)
    monthly_kpis.to_excel(writer, sheet_name="KPIs", startrow=7, startcol=0, index=False)

    # Apply header format for monthly_kpis (headers are in row 8, 0-indexed)
    for col_num, value in enumerate(monthly_kpis.columns.values):
        ws_kpis.write(7, col_num, value, header_format)

    # --- Playlist KPIs Table ---
    # Write title for Playlist KPIs
    ws_kpis.merge_range('L6:P6', 'Playlist KPIs', title_format) # L is col 11, P is col 15 (5 columns)

    # Write playlist_kpis starting from row 8 (0-indexed 2) and col L (0-indexed 11)
    playlist_kpis.to_excel(writer, sheet_name="KPIs", startrow=7, startcol=11, index=False)

    # Apply header format for playlist_kpis (headers are in row 2, 0-indexed)
    for col_num, value in enumerate(playlist_kpis.columns.values):
        ws_kpis.write(7, col_num + 11, value, header_format)

    # Optional: set column widths
    ws_data.autofit()
    ws_kpis.autofit()

    # --- Charts sheet ---
    ws_charts = wb.add_worksheet('Charts')

    # Create a new chart object for monthly data.
    chart_monthly = wb.add_chart({'type': 'line'})

    # Configure the chart series for monthly KPIs.
    max_row_monthly = len(monthly_kpis)

    chart_monthly.add_series({
        'name':       '=KPIs!$H$8',
        'categories': '=KPIs!$A$9:$A$' + str(8 + max_row_monthly),
        'values':     '=KPIs!$H$9:$H$' + str(8 + max_row_monthly),
        'marker':     {'type': 'automatic'},
    })

    chart_monthly.add_series({
        'name':       '=KPIs!$C$8',
        'categories': '=KPIs!$A$9:$A$' + str(8 + max_row_monthly),
        'values':     '=KPIs!$C$9:$C$' + str(8 + max_row_monthly),
        'marker':     {'type': 'automatic'},
    })

    # Add a chart title and some axis labels.
    chart_monthly.set_title({'name': 'Monthly Subscribers and Watch Time'})

    # Insert the chart into the worksheet.
    ws_charts.insert_chart('B2', chart_monthly)

    # Create a new chart object for playlist data.
    chart_playlist = wb.add_chart({'type': 'column'})

    # Configure the chart series for playlist KPIs.
    max_row_playlist = len(playlist_kpis)

    chart_playlist.add_series({
        'name':       '=KPIs!$N$8', # total_watch_min header (Col N, Row 8)
        'categories': '=KPIs!$L$9:$L$' + str(8 + max_row_playlist), # playlist names (Col L, Row 9 to end)
        'values':     '=KPIs!$N$9:$N$' + str(8 + max_row_playlist), # total_watch_min data (Col N, Row 9 to end)
    })



    # Add a chart title and some axis labels.
    chart_playlist.set_title({'name': 'Playlist Performance'})
    # chart_playlist.set_x_axis({'name': 'Playlist'})
    chart_playlist.set_y_axis({'name': 'Watch Time (Min)'})
    chart_playlist.set_legend({'none': True})

    # Insert the chart into the worksheet next to the monthly chart.
    ws_charts.insert_chart('J2', chart_playlist)


print(f"Excel report created: {output_file}")
```
